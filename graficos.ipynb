```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
caminho = "/content/drive/MyDrive/analise-populacao/data/populacao.csv"
df = pd.read_csv(caminho)
df.head()
```

```python
df.sort_values("População", ascending=False, inplace=True)
plt.figure(figsize=(10,6))
sns.barplot(data=df, x="População", y="País", palette="Blues_d")
plt.title("Países mais populosos")
plt.xlabel("População")
plt.ylabel("País")
plt.show()
```

```python
continentes = df.groupby("Continente")["População"].sum().sort_values(ascending=False)
continentes.plot(kind="pie", autopct="%1.1f%%", figsize=(8,8), title="População por Continente")
plt.ylabel("")
plt.show()
```

## Conclusões

- A **Ásia** lidera em população mundial, com China e Índia.
- A **Nigéria** é o país africano mais populoso.
- **Moçambique** tem população média comparada aos outros países.
